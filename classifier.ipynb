```python
import keras
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
((x_train).shape ,(y_train).shape)
```

```python
((x_test).shape , (y_test).shape)
```

```python
print(x_train[0]) # will print 28rows and 28 colom of the image and where 0 show intense white or clean white while 255 show black
print(y_train[0])
```

```python
classLabel=["Tshirt/Top",  "Trouser",  "Pullover",  "Dress",  "Coat",  "Sandal",  "Shirt",  "Sneaker",  "Bag",  "Ankle boot"]
```

```python
# giving color to the 1st image of thr train dataset
plt.imshow(x_train[0] , cmap="Greys")


# quality is not too good because of low pixels (28x28)
```

```python
plt.figure(figsize=(10,10))
for i in range(30):
    plt.subplot(5,6,i+1)
    plt.imshow(x_train[i] , cmap="Greys")
    plt.axis('off')
    plt.title(classLabel[y_train[i]])
```

FEATURE SCALING

```python
# converting 0-255 to 0-1

x_train=x_train/255
x_test=x_test/255

print(x_train[0])
```

```python
model = keras.models.Sequential(

                        [
                        keras.layers.Flatten(input_shape=(28,28)),  #convert 28x28 input matrix into single array
                        keras.layers.Dense(50,activation='relu'),
                        keras.layers.Dense(50,activation='relu'),
                        keras.layers.Dense(10,activation='softmax')
                        ]
)
```

```python
model.summary()

# param basically a number of weights and biases

# Each node in the hidden layer has weights associated with every connection made from the nodes in the input layer to that particular node in the hidden layer.
#if we have 3 node in input layer and all 3 is conneceted to the h1 node of the hidden layer then weight of the h1 is w11,w21,w31
# and if all node is connected to h2 then weight of the h2 is w21,w22,w32 so on...

# so total w= of the input layer= number of connection * number of nodes in the hidden layer + biases

```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python

model.fit(x_train,y_train,epochs=10)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)

#y_pred will indeed store an array of probabilities for each class for the corresponding x_test samples.
```

```python
print(y_pred[4].round(2))
np.argmax(y_pred[4].round(2))
```
